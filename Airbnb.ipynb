**5. data analysis**
**EDA**
**Univariate Analysis**

```python
#price distribution

df['price']
```

```python
#identifying outliers

sns.boxplot(data=df,x='price')
```

```python
#removing outliers

data_set= df[df['price']<1500]

sns.boxplot(data= data_set,x='price')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=data_set,x='price',bins=100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='availability_365',bins=30)
plt.title('Availability Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
df.columns
```

```python
#in each neighbourhood group what's avg price

avg_price = data_set.groupby('neighbourhood_group')['price'].mean()
avg_price
```

**Feature Engineering**

```python
# new column price_per_bed

data_set['price_per_bed']=df['price']/df['beds']
data_set.head()
```

```python
#in each neighbourhood group what's avg price per bed

avg_price_per_bed= data_set.groupby('neighbourhood_group')['price_per_bed'].mean()
avg_price_per_bed
```

**Bi Variate Analysis** - To check one variable dependency on another variable

```python
data_set.columns
```

```python
#how price is dependent on neighbourhood


sns.barplot(data=data_set,x='neighbourhood_group',y='price',hue='room_type')
plt.title('Neighbourhood by Price and room type')
plt.show()
```

```python
#review and price relation
sns.scatterplot(data=data_set,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.title('Locality and Review Dependency')
plt.show()
```

```python
data_set.info()
```

```python
sns.pairplot(data=data_set, vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
```

```python
#geographical distribution with price
plt.figure(figsize=(10,7))
sns.scatterplot(data=data_set,x='longitude',y='latitude', hue='price')
```

```python
#geographical distribution with room type
plt.figure(figsize=(10,7))
plt.title('Geographical Distribution of AirBNB Listings')
sns.scatterplot(data=data_set,x='longitude',y='latitude',hue='room_type')
plt.show()
```
